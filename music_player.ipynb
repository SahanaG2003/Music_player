```python
import os 
import pygame
from pygame import mixer
from tkinter import Tk, ttk
from tkinter import PhotoImage 
from tkinter import filedialog
from tkinter import Button
from tkinter import Frame
from tkinter import Label
from tkinter import Canvas
from tkinter import Scale
from pydub import AudioSegment
from PIL import Image 
from PIL import ImageTk
from mp3_tagger import MP3File
import time
```

```python
file_path = None
is_playing = False
start_time = 0

def choose_file():
    global file_path
    if is_playing:
        stop_music()
    file_path = filedialog.askopenfilename()
    play_music()

# Function to play music
def play_music():
    global is_playing, start_time
    pygame.mixer.init()
    pygame.mixer.music.load(file_path)
    pygame.mixer.music.play()
    current_song = os.path.basename(file_path)
    is_playing = True

    # Update the song information label
    top_song_info_label.config(text=f"{current_song}")

    # Set the start time
    start_time = time.time()

    # Start updating the elapsed time label
    update_elapsed_time()

def stop_music():
    global is_playing
    pygame.mixer.music.stop()

    # Stop updating the elapsed time label
    is_playing = False

    # Update the song information label
    top_song_info_label.config(text="No Song Playing")

def pause_music():
    pygame.mixer.music.pause()

def resume_music():
    pygame.mixer.music.unpause()

def next_song():
    stop_music()
    # Load the next song in the playlist (you need to implement your playlist logic)
    choose_file()

def change_volume(val):
    volume = int(val) / 100.0
    if not pygame.mixer.get_init():
        pygame.mixer.init()
    pygame.mixer.music.set_volume(volume)

def update_elapsed_time():
    if is_playing:
        elapsed_time = time.time() - start_time
        elapsed_minutes = int(elapsed_time / 60)
        elapsed_seconds = int(elapsed_time % 60)
        elapsed_time_label.config(text=f"Time: {elapsed_minutes:02d}:{elapsed_seconds:02d}")
        root.after(1000, update_elapsed_time)  # Schedule the function to be called after 1000 milliseconds (1 second)
```

```python
root = Tk()
root.title("MUSIC PLAYER")
root.geometry("485x650+290+10")
root.configure(background="#999999")
root.resizable(False, False)

# Create a background canvas
bg_canvas = Canvas(root, width=485, height=500, highlightthickness=0)
bg_canvas.pack()

# Load and display the background image
bg_image = ImageTk.PhotoImage(file="music.gif")
bg_canvas.create_image(0, 0, anchor="nw", image=bg_image)

# Create labels for song information and elapsed time
top_song_info_label = Label(root, text="No Song Playing", font=("Helvetica", 12), background="#444444", fg="white")
top_song_info_label.pack(side="top", fill="both", expand=False, padx=10, pady=5)

elapsed_time_label = Label(root, text="Time: 00:00", font=("Helvetica", 12), bg='#444444', fg="white")
elapsed_time_label.pack(side="top", fill="both", expand=True, padx=10, pady=5)

# Create a lower frame for buttons and controls
lower_frame = Frame(root, bg='#444444', width=485, height=200)
lower_frame.pack(side="bottom")

# Create buttons for controlling music
select_button = Button(lower_frame, text="Browse", command=choose_file)
select_button.pack(side="left", pady=7, padx=7)

next_button = Button(lower_frame, text="Next Song", command=next_song)
next_button.pack(side="left", pady=7, padx=7)

play_button = Button(lower_frame, text="Play", command=play_music)
play_button.pack(side="left", pady=7, padx=7)

stop_button = Button(lower_frame, text="Stop", command=stop_music)
stop_button.pack(side="left", pady=7, padx=7)

pause_button = Button(lower_frame, text="Pause", command=pause_music)
pause_button.pack(side="left", pady=7, padx=7)

resume_button = Button(lower_frame, text="Resume", command=resume_music)
resume_button.pack(side="left", pady=7, padx=7)

volume_scale = Scale(lower_frame, from_=0, to=100, orient="horizontal", command=change_volume)
volume_scale.set(50)  # Set initial volume to 50%
volume_scale.pack(side="right", pady=7, padx=7)

# Start the Tkinter event loop
root.mainloop()
```
